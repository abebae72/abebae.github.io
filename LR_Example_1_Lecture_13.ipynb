```python
# Linear Regression Example 1:    A Single Predictor Variable Case   Y = w0 + w1*X
```

```python
from sklearn.linear_model import LinearRegression        # sklearn is a library for "Linear Regression" in Python.
                                                         # linear_model is one of sub-library.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
weight = np.array([[60], [60.5], [72.3], [75], [80]])
```

```python
height = np.array([[160], [165], [170], [173], [180]])
```

```python
plt.xlabel("Weight")
plt.ylabel("Height")
plt.axis([0, 100, 0, 190])         # First Two Digits: Range of Horizontal Axis,  Latter Two Digits: Vertical
plt.scatter(weight, height)
```

```python
model = LinearRegression()       # LinearRegression() is sub-library for "linear_model"
```

```python
model.fit(weight, height)                     # Execution of Linear Regression
```

```python
w1 = model.coef_                              # Coefficient w1
w0 = model.intercept_                         # Intercept w0
print("w1 = ", w1)
print("w0 = ", w0)
```

```python
regression_height = w0 + w1 *weight           # Model Prediction
plt.axis([0, 140, 0, 220])                  # Axis Range Expansion
plt.plot(weight, regression_height)           # Plotting of Model
```

```python
plt.xlabel("Weight")
plt.ylabel("Height")
plt.axis([0, 140, 0, 220])
plt.scatter(weight, height)
overall_weight = [[0], [120]]                
plt.plot(overall_weight, model.predict(overall_weight), color='blue')      # Plotting of Model for Overall Weights
```

```python
from sklearn.metrics import mean_squared_error      # mean_squared_error is a sub-library of sklearn.metrics
```

```python
predicted_height = model.predict(weight)
mse = mean_squared_error(height, predicted_height)   # mean_squared_error = sum of (height - predicted_height)^2
```

```python
print("Mean Squared Error =", mse)
```
