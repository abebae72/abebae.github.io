```python
# Lists (리스트)    : Element Modification
```

```python
motors = ['benz', 'bmw', 'toyoda', 'hyundai']
print(motors)
```

```python
motors[2] = 'honda'
print(motors)
```

```python
motors.append('toyota')            # append()
print(motors)
```

```python
motors.insert(0, 'ford')            # insert()
print(motors)
```

```python
motors.sort()                   # sort()  : alphabetical order
print(motors)
```

```python
motors.sort(reverse=True)                   
print(motors)
```

```python
motors.reverse()                   
print(motors)
```

```python
digits = list(range(1,6))            # list() and range()
print(digits)
```

```python
digits_1 = list(range(2, 11, 2))            # beging with "2", and consecutively add "2" until "11" or passing "11"
print(digits_1)
```

```python
digits_2 = list(range(0, 10))            #  various commands, max(), min(), sum()
print(max(digits_2))
print(min(digits_2))
print(sum(digits_2))
```

```python
len(motors)                  #  len() : the size (or length) of the list
```

```python
cars = ['benz', 'bmw', 'ford', 'honda', 'hyundai', 'toyota']
print(cars[0:3])
print(cars[1:4])
print(cars[:4])
print(cars[2:])
```
