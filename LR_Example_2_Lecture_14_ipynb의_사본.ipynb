```python
# Linear Regression Example 2:    A Single Variable (X) Case  vs. Multiple Variables (X, Z) Case   (Y = w0 + w1*X  vs. Y = w0 + w1*X + w2*X*X)
```

```python
# A Single Variable (X) Case  (Y = w0 + w1*X)
```

```python
import numpy as np
```

```python
perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])
```

```python
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("Length")
plt.ylabel("Weight")
plt.scatter(perch_length, perch_weight)
```

```python
length = perch_length.reshape(-1,1)
weight = perch_weight.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
model = LinearRegression()   
```

```python
model.fit(length, weight)   
```

```python
w1 = model.coef_                              # Coefficient w1
w0 = model.intercept_                         # Intercept w0
print("w1 = ", w1)
print("w0 = ", w0)
```

```python
regression_weight = w0 + w1 * length           # Model Prediction
plt.axis([0, 50, -800, 1200])                  # Axis Range Expansion
plt.plot(length, regression_weight)            # Plotting of Model
```

```python
plt.xlabel("Length")
plt.ylabel("Weight")
plt.axis([0, 50, -800, 1200]) 
plt.scatter(length, weight)
overall_length = [[0], [50]]                
plt.plot(overall_length, model.predict(overall_length), color='blue')      # Plotting of Model for Overall Weights
```

```python
from sklearn.metrics import mean_squared_error       # mean_squared_error is a sub-library of sklearn.metrics
```

```python
predicted_weight = model.predict(length)
mse = mean_squared_error(weight, predicted_weight)   # mean_squared_error = sum of (height - predicted_height)^2
```

```python
print("Mean Squared Error =", mse)
```

```python
# Multiple Variables (X, Z) Case   (Y = w0 + w1*X + w2*X*X)
```

```python
poly_length = np.column_stack((length**2, length))   # creating of anothre column for (length * length) in addition to existing column (length)
```

```python
model_poly = LinearRegression()   
```

```python
model_poly.fit(poly_length, weight)   
```

```python
w = model_poly.coef_                 # Coefficient w2, w1
w0 = model_poly.intercept_            # Intercept w0
print("w2, w1 = ", w)
print("w0 = ", w0)
```

```python
ws = w.reshape(-1, 1) 
```

```python
ws
```

```python
w2 = ws[0]
w1 = ws[1]
```

```python
plt.xlabel("Length")
plt.ylabel("Weight")
plt.axis([0, 50, -800, 1200]) 
plt.scatter(length, weight)
poly_regression_weight = w0 + w1*length + w2*length*length          # Model Prediction
plt.plot(length, poly_regression_weight)                            # Plotting of Model
```

```python
predicted_weight = model_poly.predict(poly_length)
mse = mean_squared_error(weight, predicted_weight)   # mean_squared_error = sum of (height - predicted_height)^2
```

```python
print("Mean Squared Error =", mse)
```
