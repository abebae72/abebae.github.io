```python
# Matplotlib Calling
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# matplotlib : linespacing() -> a total of "50" evenly-spaced data is created !
```

```python
x = np.linspace (-5, 5)     
print(x)
print(len(x))
```

```python
y = 2 * x
plt.plot(x, y)                     #  plt.plot()
plt.show()                         #  plt.show()
```

```python
# matplotlib : axis expression in plot
```

```python
x = np.linspace(-5, 5)
y_1 = 2 * x
y_2 = 3 * x
```

```python
plt.xlabel("x value")                        # x-axis
plt.ylabel("y value")                        # y-axis
plt.title("First Graph in Python")           # graph title
```

```python
plt.plot(x, y_1, label="y1")                         # legend naming
plt.plot(y, y_2, label="y2", linestyle="dashed")     # line style configuration
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#  Pandas : Series()     One Dimensional Data     
#           DataFrame()  Two Dimensional Data
```

```python
# Pandas : Series Creating
```

```python
a = pd.Series([60,80,70,20,30], index=["Korean", "Math", "English", "War History", "AI Civil"])      # "index" allocates the series data to label value!
print(type(a))
print(a)
```

```python
# Pandas : Series Creating with NumPy array()
```

```python
a = pd.Series(np.array([60,80,70,20,30]), index = np.array(["Korean", "Math", "English", "War History", "AI Civil"]))
print(type(a))
print(a)
```

```python
# Pandas : Series Creating with Direct Assigning
```

```python
a = pd.Series({"Korean":60, "Math":80, "English":70, "War History":20, "AI Civil":30})
print(type(a))
print(a)
```

```python
# Pandas : Series Manipulation
```

```python
a = pd.Series({"Korean":60, "Math":80, "English":70, "War History":20, "AI Civil":30})
print(a[2])
print(a["AI Civil"])
```

```python
a = pd.Series({"Korean":60, "Math":80, "English":70, "War History":20, "AI Civil":30})      
b = pd.Series([100],index=["Shooting"])
a = a.append(b)                                  # append()                          
print(a)
```
